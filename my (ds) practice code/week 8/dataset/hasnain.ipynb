```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =  pd.read_csv(r"C:\Users\dilib.DESKTOP-8O9VNIH\Desktop\hasnain data\dataset\BankChurners.csv")
```

```python
data['Marital_Status']
```

```python
data_cleaned = data.iloc[:, :-2]
data_cleaned
```

```python
sample_data = data_cleaned.sample(n=700, random_state=42)
sample_data
```

```python
marital_status_counts = sample_data['Marital_Status'].value_counts()
marital_status_counts
```

```python
sns.set(style="whitegrid")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Analysis of Sampled Bank Data', fontsize=16)
```

```python
sns.countplot(data=sample_data, x='Marital_Status', hue='Marital_Status', palette="Set2", ax=axes[0, 0])
legend = axes[0, 0].get_legend()
print("", legend)
if legend is not None:
    legend.remove() 
axes[0, 0].set_title('Marital Status Distribution')
```

```python
sns.histplot(data=sample_data, x='Customer_Age', hue='Marital_Status', multiple="stack", palette="Set3", ax=axes[0, 1])
axes[0, 1].set_title('Age Distribution by Marital Status')
```

```python
sns.countplot(data=sample_data, x='Income_Category', hue='Marital_Status', palette="Set1", ax=axes[1, 0])
axes[1, 0].set_title('Income Distribution by Marital Status')
axes[1, 0].tick_params(axis='x', rotation=45)
```

```python
# Average Credit Limit by Marital Status (adjusted for warning)
credit_limit_mean = sample_data.groupby('Marital_Status')['Credit_Limit'].mean().reset_index()
sns.barplot(data=credit_limit_mean, x='Marital_Status', y='Credit_Limit', ax=axes[1, 1], color='skyblue')
axes[1, 1].set_title('Average Credit Limit by Marital Status')
```

```python
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
