```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
X= np.sort(5*np.random.rand(80, 1),axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
print(f'Training Error: {mean_squared_error(y_train, y_train_pred)}')
print(f'Testing Error: {mean_squared_error(y_test, y_test_pred)}')
```

```python
# Step 4: Plot the training data, the true function, and the model's predictions
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Testing Data')
plt.plot(X, np.sin(X), color='red', label='True Function')
plt.plot(X, model.predict(X), color='orange', label='Model Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Underfitting Example')
plt.legend()
plt.show()



```

```python
from sklearn.tree import DecisionTreeRegressor

# Fit a Decision Tree model
model = DecisionTreeRegressor(max_depth=15)  # High depth to illustrate overfitting
model.fit(X, y)
```

```python
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Overfitted Tree')
plt.title('Example of Overfitting')
plt.legend()
plt.show()
```
